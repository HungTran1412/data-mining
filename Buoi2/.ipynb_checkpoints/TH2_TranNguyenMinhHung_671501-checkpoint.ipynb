### Data Manipulation: làm việc với dữ liệu

```python
df.item_price.dtype
```

#convert kiểu dữ liệu

### Apply()

```python
#đổi kiểu dữ liệu và
#gán lại vào cột item_price sau khi đã đổi kiểu dữ liệu
df.item_price = df.item_price.apply(lambda x: float(x.replace('$', '')))
```

```python
df.head()
```

```python
# tạo cột mới để tính tổng tiền
df["total_price"] = df["quantity"] * df["item_price"]
df.head()
```

### How much was the revenue for the period in the dataset?
tổng doanh thu ở trong file

```python
revenue = df.total_price.sum()
print("Tổng doanh thu là: ", revenue)
```

## Group by

```python
# gom nhosm theo so luong
c = df.groupby("item_name")["quantity"].sum()
#sap xep giam dan va dau ra 5 mon ban chay nhat
c.sort_values(ascending = False).head(5)
```

```python
#Unique Value, tìm tất cả món hàng khác nhau
df.item_name.value_counts().count()
```
